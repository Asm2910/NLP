```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
data = pd.read_csv('data.csv', error_bad_lines=False)
data.head()
```

```python
data['strength'].unique()
```

```python
data.isna().sum()
```

```python
data[data['password'].isnull()]
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
sns.countplot(data['strength'])
```

```python
pt = np.array(data)
pt
```

```python
import random
random.shuffle(pt)
```

```python
x = [labels[0] for labels in pt]
y = [labels[1] for labels in pt]
```

```python
x
```

```python
y
```

```python
#function to split input into char of list

def word_divide(inputs):
    character =[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide('kzde5577')
```

```python
#import TF-IDF vectorizer to convert String data into numerical data

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector = TfidfVectorizer(tokenizer= word_divide)
```

```python
#apply tfidf vectorizer on data

X = vector.fit_transform(x)
X.shape
```

```python
vector.get_feature_names()
```

```python
first_document_vector = X[0]
first_document_vector 
```

```python
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vector.get_feature_names(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train, y_train)
```

```python
#doing prediction for specific custom data
dt=np.array(['%@123abcd'])
pred=vector.transform(dt)
clf.predict(pred)
```

```python
#on test data
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
